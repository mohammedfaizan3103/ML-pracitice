Q2. Write a Python program using Scikit-learn to print the keys, number of rows-columns,
feature names and the description of the Iris data.

```python
from sklearn import datasets
iris = datasets.load_iris()

print("Keys:", iris.keys())
print("Rows:", iris.data.shape[0])
print("Columns:", iris.data.shape[1])
print("Feature names:", iris.feature_names)
print("Description of the dataset:", iris.DESCR)
```

Q3. Write a Python program to get the number of observations, missing values and nan values.

```python
import sklearn
import numpy as np
from sklearn import datasets
iris = datasets.load_iris()
print("no.of observations: ", iris.data.shape[0])
missingvals = np.isnan(iris.data).sum()
print("no.of missing values: ", missingvals)
```

Q7. Write a Python program to access first four cells from a given Dataframe using the index and column labels. Call iris.csv to create the Dataframe.

Q8. Write a Python program to create a plot to get a general Statistics of Iris data.

```python
import matplotlib.pyplot as plt

stats = df.describe()
stats[1:].plot(kind='bar')
plt.title('General Statistics of IRIS Data')
plt.tight_layout()
plt.show()
```

Q9. Write a Python program to create a Bar plot to get the frequency of the three species of the Iris data.

```python
import seaborn as sns
sns.countplot(x='variety', data=df)
plt.title('Species Frequency')
plt.show()
```

Q10. Write a Python program to create a Pie plot to get the frequency of the three species of the Iris data.

```python
df['variety'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Species Distribution')
plt.ylabel('')
plt.show()
```

Q11. Write a Python program to create a graph to find relationship between the sepal length and width.

```python
df.plot.scatter(x='sepal.length', y='sepal.width', c= 'Orange')
plt.show()
```

Q12. Write a Python program to create a graph to find relationship between the petal length

Q13. Write a Python program to create a graph to see how the length and width of SepalLength, SepalWidth, PetalLength, PetalWidth are distributed.

```python
df[['sepal.length','sepal.width','petal.length','petal.width']].hist(figsize=(10,8))
plt.show()
```

Q14. Write a Python program to create a joinplot to describe individual distributions on the same plot between Sepal length and Sepal width.

```python
sns.jointplot(
    x='sepal.length',
    y='sepal.width',
    data=df,
    color='blue'
)
plt.show()
```

Q15. Write a Python program to create a joinplot using "hexbin" to describe individual distributions on the same plot between Sepal length and Sepal width.

```python
sns.jointplot(
    x='sepal.length',
    y='sepal.width',
    data=df,
    kind='hex',
    color='green'
)
plt.show()
```
